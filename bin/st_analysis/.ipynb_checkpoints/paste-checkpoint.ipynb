```python
import pandas as pd
import anndata as ad
import numpy as np
import os
import re
import seaborn as sns
import argparse
import scanpy as sc
from matplotlib import cm
import matplotlib.pyplot as plt
from src.classes import *
import paste as pst
import ot
import time
from matplotlib.pyplot import figure

```

```python
path = '../../data/st_data_pre_processed/stdata_h5ad/'
slides = ['ST3_C2', 'ST3_D2', 'ST3_E1','ST3_E2', 'ST3_D1']
files = [path + x for x in os.listdir(path) if x.split('_st')[0] in slides and x.endswith('_stdata.h5ad')]
files = [x for x in files if 'CN56_C1' not in x]
print(files)
sample_list=[]
for file in files:
    adata = sc.read_h5ad(file)
    sc.pp.filter_genes(adata, min_counts=15)
    sc.pp.filter_cells(adata, min_counts=100)
    sc.pp.calculate_qc_metrics(adata, inplace=True)
    sample_list.append(adata)
```

```python
# align the slices with PASTE - center

initial_slice = sample_list[0]
lmbda = len(sample_list)*[1/len(sample_list)]
```

```python
pst.filter_for_common_genes(sample_list)

b = []
for i in range(len(sample_list)):
    print(f'Comparing sample {sample_list[i].uns["sample"]} to the initial slide ({i+1}/{len(sample_list)})')
    b.append(pst.match_spots_using_spatial_heuristic(sample_list[0].X, sample_list[i].X))
```

```python
start = time.time()

center_slice, pis = pst.center_align(initial_slice, 
                                     sample_list, 
                                     lmbda,
                                     random_seed=5, 
                                     pis_init = b, 
                                     alpha=0.5,
                                     n_components=40)

print('Runtime: ' + str(time.time() - start))
```

```python
center, new_slices = pst.stack_slices_center(center_slice, sample_list, pis)
```

```python
new_slices[1].obs
```

```python
# x = len(sa)
gene='PENK'
fig, axs = plt.subplots(2, 2,figsize=(7,10.5))
axs[0,0].scatter(center.obsm['spatial'][:, 0],
            center.obsm['spatial'][:, 1],
            c = center[:, gene].X.flatten().tolist(), s=20)

axs[0,1].scatter(new_slices[1].obsm['spatial'][:, 0],
            new_slices[1].obsm['spatial'][:, 1],
            c = new_slices[1][:, gene].X.flatten().tolist(), s=20)

axs[1,0].scatter(new_slices[2].obsm['spatial'][:, 0],
            new_slices[2].obsm['spatial'][:, 1],
            c = new_slices[2][:, gene].X.flatten().tolist(), s=20)

axs[1,1].scatter(new_slices[0].obsm['spatial'][:, 0],
            new_slices[0].obsm['spatial'][:, 1],
            c = new_slices[0][:, gene].X.flatten().tolist(), s=20)
plt.show()
```

```python
# processing according to the new matrix X
center_slice_exp = center_slice.copy()

# normalize and log transform
sc.pp.normalize_total(center_slice_exp, target_sum=10000)
sc.pp.log1p(center_slice_exp)

# select and plot HVG
sc.pp.highly_variable_genes(center_slice_exp, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(center_slice_exp)

# save log transformed expression for later
center_slice_exp.raw = center_slice_exp.copy()

# use the regress out function
sc.pp.regress_out(center_slice_exp, ['n_counts', 'MT_perc'])
sc.pp.scale(center_slice_exp, max_value=10)

sc.tl.pca(center_slice_exp, svd_solver='arpack')
sc.pl.pca(center_slice_exp, color='TH')
sc.pl.pca_variance_ratio(center_slice_exp)

# compute neighbours and cluster
sc.pp.neighbors(center_slice_exp, n_neighbors=10, n_pcs=14)

sc.tl.umap(center_slice_exp)
sc.pl.umap(center_slice_exp, color=['TH', 'RET', 'PENK'], use_raw=False)

sc.tl.leiden(center_slice_exp, resolution=1)
sc.pl.umap(center_slice_exp, color=['leiden', 'TH', 'PENK'])

# plot clusters onto the central slide
figure(figsize=(6, 8), dpi=50)
plt.scatter(x = center_slice_exp.obsm['spatial'][:, 0],y= center_slice_exp.obsm['spatial'][:, 1], 
            c=center_slice_exp.obs['leiden'].astype('int'), cmap='jet', s=55)
plt.colorbar()
plt.title('Center slide cluster assignment')
plt.axis('off')
plt.show()

# calculate DEG
sc.tl.rank_genes_groups(center_slice_exp, 'leiden', method='t-test')
sc.pl.rank_genes_groups(center_slice_exp, n_genes=25, sharey=False)
```

```python
# use the paste_W matrix
center_slice_exp = center_slice.copy()
center_slice_exp.obsm['X_paste'] = center_slice_exp.uns['paste_W']
sc.pp.neighbors(center_slice_exp, use_rep='X_paste')
sc.tl.leiden(center_slice_exp, resolution=1.25, key_added='leiden_1.25')
```

```python
figure(figsize=(6, 8), dpi=50)
plt.scatter(x = center_slice_exp.obsm['spatial'][:, 0],y= center_slice_exp.obsm['spatial'][:, 1], 
            c=center_slice_exp.obs['leiden_1.25'].astype('int'), cmap='jet', s=55)
plt.colorbar()
plt.title('Center slide cluster assignment')
plt.axis('off')
plt.show()
```
